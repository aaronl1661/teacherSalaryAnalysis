```python
df = pd.read_csv('./salary_data/employee_data.csv')

combine_df = pd.DataFrame()

curr = ["", "", "", ""]
count = 0
partial_matches = []
matches = []

def intersection(l1, l2):
    output = []
    temp1 = l1[1:-1]
    temp2 = l2[1:-1]
    list1 = list(temp1.split(", "))
    list2 = list(temp2.split(", "))
    for i in list1:
        if i in list2:
            output.append(i)
    return output

for index, row in df.iterrows():
    #print(df.at[index, "name"])
    if curr[0] == df.at[index, "name"]:
        print(df.at[index, "name"])
        # x = intersection(curr[1], df.at[index, "years_taught"])
        # print(x)
        #print(list(set(curr[1]) & set(df.at[index, "years_taught"])))
        #print(list(set(curr[3]) & set(df.at[index, "departments"])))
        #if list(set(curr[1]) & set(df.at[index, "years_taught"])) == [] and list(set(curr[3]) & set(df.at[index, "departments"])) != []:
        if intersection(curr[1], df.at[index, "years_taught"]) == [] and intersection(curr[3], df.at[index, "departments"]) != []:
            #print("A")
            curr[1] = curr[1][:-1] + (df.at[index, "years_taught"])[1:]
            curr[3] = curr[3][:-1] + (df.at[index, "departments"])[1:]
            matches.append(df.at[index, "name"])
            if index == len(df.index) -1:
                combine_df.at[count, "name"] = curr[0]
                combine_df.at[count, "years_taught"] = curr[1]
                combine_df.at[count, "salaries"] = curr[2]
                combine_df.at[count, "departments"] = curr[3]
                curr = [df.at[index, "name"], df.at[index, "years_taught"], df.at[index, "salaries"], df.at[index, "departments"]]
                count = count + 1
        else:
            partial_matches.append(df.at[index, "name"])
            combine_df.at[count, "name"] = curr[0]
            combine_df.at[count, "years_taught"] = curr[1]
            combine_df.at[count, "salaries"] = curr[2]
            combine_df.at[count, "departments"] = curr[3]
            curr = [df.at[index, "name"], df.at[index, "years_taught"], df.at[index, "salaries"], df.at[index, "departments"]]
            count = count + 1
    else:
        combine_df.at[count, "name"] = curr[0]
        combine_df.at[count, "years_taught"] = curr[1]
        combine_df.at[count, "salaries"] = curr[2]
        combine_df.at[count, "departments"] = curr[3]
        curr = [df.at[index, "name"], df.at[index, "years_taught"], df.at[index, "salaries"], df.at[index, "departments"]]
        count = count + 1

combine_df = combine_df.drop(labels = 0, axis = 0)

combine_df
```
