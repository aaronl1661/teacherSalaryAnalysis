```python
departments = {}

with open("departments.txt", "r") as file:
    lines = file.readlines()
    for i, line in enumerate(lines):
        if line == "\n" and i < len(lines)-3:
            dep = lines[i+2].strip()
            departments[dep] = []
        else:
            if lines[i-1] != "\n":
                departments[dep].append(line.strip())
print(departments)
file.close()

matched = {}

with open('department_majors.txt', 'r') as f:
    lines = f.readlines()
    for line in lines:
        line = line.strip()
        major = line[5:]
        m = line[:4]
        #print(major)
        found = 0
        for dep in departments:
            #print(departments[dep])
            for d in departments[dep]:
                if major in d:
                    if found == 0:
                        #print(f"matched: {major} to department {dep}")
                        found = 1
                        matched[m] = dep
                    
        if not found:
            print(major)

print(matched)
```
