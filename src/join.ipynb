```python
import pandas as pd

gpa_review_df = pd.read_csv('./grades_reviews/cleaned_data.csv')
salary_df = pd.read_csv('./salary_data/employee_data.csv')
# has Employee column and name column

print(f'professors: {len(gpa_review_df)}, paid employees: {len(salary_df)}')

outer = pd.merge(
    gpa_review_df,
    salary_df,
    how = "outer",
    on = "name",
    indicator=True
)

inner = pd.merge(
    gpa_review_df,
    salary_df,
    how = "inner",
    on = "name",
    indicator=True
)

left = pd.merge(
    gpa_review_df,
    salary_df,
    how = "left",
    on = "name",
    indicator=True
)

print(f'new outer len: {len(outer)}, new inner len: {len(inner)},  new left len: {len(left)}')

#not_in_merge = gpa_review_df[~gpa_review_df.index.isin(both_df.index)]
left_only = outer[outer['_merge'] == 'left_only']
print(len(left_only))
both = outer[outer['_merge'] == 'both']
print(len(both))

print(left_only.to_string())
inner.to_csv(f'combined_data.csv', index=False)
```
